```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

# I import my package:

import averagedistorsion as ad

font = {'size': 14}
  
# using rc function
plt.rc('font', **font)
```

The questions here are : (1) who chose the positions of the relevant alternatives $a$ and $b$? and (2) who chose the positions of the irrelevant alternatives $c_1, \dots, c_m$ depending on $a$ and $b$? For each question we can answer independently (a) The player (b) The nature (c) An adversarial opponent. We do not consider case (c) here but we consider (a) + (a), (a) + (b), (b) + (a) and (b) + (b).

```python
rules = [#ad.RuleAlwaysWorst(),
         #ad.RuleLottery(), 
         #ad.RuleDictatorRandom(),
         #ad.RuleEgalitarian(),
         #ad.RuleVeto(), 
         ad.RulePlurality(), 
         ad.RulePluralityVeto(),
         ad.RuleSTV(), 
         ad.RuleHarmonic(), 
         ad.RuleHalfApproval(), 
         ad.RuleRankedPairs(),
         ad.RuleCopeland(),
         #ad.RuleMedian(), 
         ad.RuleBorda(), 
         ad.RuleOddBorda(),
         #ad.RuleRandomVoterUtility(),
         #ad.RuleNashProduct()
         ]

n_candidates = 2
n_irrelevant_candidates = 13
n_voters = 30
n_tries = 10000
```

```python
experiment = ad.ExperimentIrrelevantCandidates(rules, 
                                                n_candidates=n_candidates,
                                                n_irrelevant_candidates=n_irrelevant_candidates,
                                                n_voters=n_voters,
                                                n_tries=n_tries)
```

# 1. Nature chose everything

## Euclidean : Uniform distribution (1D)

```python
experiment(ad.ModelEuclideanUniform(dim=1))
experiment.show_distortion("Euclidean Uniform 1D")
experiment.show_accuracy("Euclidean Uniform 1D")
experiment.save_results("results/irrelevant_experiment_euclidean_uniform_1d.csv")
```
