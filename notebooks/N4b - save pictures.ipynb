```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import tikzplotlib

# I import my package:

import averagedistorsion as ad

font = {'size': 16}
  
# using rc function
plt.rc('font', **font)
```

```python
models = ["uniform", 
          "uniform_normalized", 
          "gaussian_neutral",
          "gaussian_biaised",
          "restaurants",
          "bars",
          "politics"]
```

# Partie 1 

```python
model_dict = {}
for model in models:
    res = np.loadtxt("results/irrelevant_experiment_%s.csv"%model, delimiter=",", dtype=str)
    model_dict[model] = {}
    for row in res:
        rule = row[0]
        n_cand = int(row[2])
        dist = float(row[5])
        acc = float(row[6])
        if rule not in model_dict[model]:
            model_dict[model][rule] = {"dist":[], "acc":[]}
        model_dict[model][rule]["dist"].append(dist)
        model_dict[model][rule]["acc"].append(acc)
```

```python
for model in model_dict:
    plt.figure(figsize=(20, 10))
    for j,rule in enumerate(model_dict[model]):
        line_format = "o-"
        if j >= 10:
            line_format = "o--"
        plt.plot(np.arange(0,9), model_dict[model][rule]["dist"], line_format, linewidth=5, markersize=10, label=rule)
    plt.legend()
    plt.xlim(0,8)
    plt.ylim(1)
    plt.title("Utilities : " + model)
    plt.xlabel("Number of irrelevant alternatives")
    plt.ylabel("Distortion")
    tikzplotlib.save("Figures/irrelevant_model_%s.tex"%(model))
    plt.savefig("Figures/irrelevant_model_%s.png"%(model))
    plt.show()
```
